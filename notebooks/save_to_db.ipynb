```python
from pymongo import MongoClient
import pickle
import os
```

```python
from sys import platform as _platform

def start_platform_based_mongo_db():
    if _platform == "linux" or _platform == "linux2":
        print('This is linux.')
    elif _platform == "darwin":
        print('I ended up sending up MongoDB a different way on my Mac.')
    elif _platform == "win32":
        print('This is Windows 32-bit.')
    elif _platform == "win64":
        print('This is Windows 64-bit.')
        os.startfile(r"C:\Program Files\MongoDB\Server\4.2\bin\mongo.exe")
    
start_platform_based_mongo_db()
```

```python
def get_absolute_path(rel_path):
    parent_directory = os.getcwd() 
    file_path = os.path.join(parent_directory, rel_path)
    return file_path 

f_path = get_absolute_path('../data/pitt_data.txt')
pitt_data_csv = get_absolute_path('../data/pitt_data_file.csv')
```

```python
with open(f_path, "rb") as fp:
    data=pickle.load(fp)
```
